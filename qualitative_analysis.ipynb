```python
import pandas as pd
```

## Copilot 

```python
df = pd.read_csv(qual_file)
```

```python
df
```

```python
qual = {
    'Correct Number of Panels':{'incorrect':0, 'correct':0, 'df label':"Correct Number of Panels"},
    'Correct Plot Type':{'incorrect':0, 'correct':0, 'df label':"Correct plot type"},
    'Correct Number of Lines/Bars':{'incorrect':0, 'correct':0, 'df label':"Correct Number of Lines/Bars (NM=not mentioned)"},
    'Correct title':{'incorrect':0, 'correct':0, 'df label':"Title mentioned/Title correct (M=mentioned, C=correct, NC=not correct)"},
    'Correct x-axis labels':{'incorrect':0, 'correct':0, 'df label':"x-axis label mentioned"},
    'Correct y-axis labels':{'incorrect':0, 'correct':0, 'df label':"y-axis label mentioned (M=mentioned, C=correct, NC=not correct)"},
    'Correct x-axis tick labels':{'incorrect':0, 'correct':0, 'df label':"x-axis tick labels correct"},
    'Correct y-axis tick labels':{'incorrect':0, 'correct':0, 'df label':"y-axis tick labels correct"},
    'Error bars mentioned (when present)':{'incorrect':0, 'correct':0, 'df label':"error bars mentioned (NA means no error bars)"},
       }

for k,v in qual.items():
    d = df[v['df label']].values
    for dd in d: # all values
        if '/' not in str(dd):
            if str(dd).strip()=='Y':
                v['correct'] += 1
            elif str(dd).strip()=='N':
                v['incorrect'] += 1
            elif str(dd).strip()=='NM': # not mentioned
                v['incorrect'] += 1
            elif str(dd).strip()=='M-NC': # mentioned, but not correct
                v['incorrect'] += 1
            elif str(dd).strip()=='M-C': # mentioned, and correct
                v['correct'] += 1
            else:
                print('other:', str(dd).strip())
        else: # has multi
            d2 = dd.split('/')
            for dd2 in d2:
                if str(dd2).strip()=='Y':
                    v['correct'] += 1
                elif str(dd2).strip()=='N':
                    v['incorrect'] += 1
                elif str(dd2).strip()=='NM': # not mentioned
                    v['incorrect'] += 1
                elif str(dd2).strip()=='M-NC': # mentioned but not correct
                    v['incorrect'] += 1
                elif str(dd2).strip()=='M-C': # mentioned and correct
                    v['correct'] += 1
                else:
                    print('other, 2:', str(dd2).strip())
                
    #import sys; sys.exit()
```

```python
qual_per = {} # percent correct
for k,v in qual.items():
    qual_per[k] = v['correct']/(v['correct']+v['incorrect'])*100
```

```python
qual_per
```

Count hallucinations:

```python
hallucinations = 0
#hall_per_page = []
for d in df['other hallucinations'].values:
    if '/' not in d:
        hallucinations += int(float(d))
    else:
        dd = d.split('/')
        for d2 in dd:
            hallucinations += int(float(d2))

print('Total number of hallucinations:', hallucinations)
```
